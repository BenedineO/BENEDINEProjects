```python
# Importing pandas for data manipulation and linearRegression for model training
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Creating a variable to store the csv file
data = pd.read_csv("house_prices.csv")
```

```python
# Viewing the dataset
data
```

```python
# Spliting the data into features (X) input and (y) output
X = data [['Size', 'Location']]
y = data ['Price']
```

```python
# Creating the model using linear Regression
model = LinearRegression()
```

```python
# Training the model
model.fit(X, y)
```

```python
print(X)
```

```python
prediction = model.predict ([[1200,1]])
```

```python
print(f'Predicted price for a house with size 1200 in location 1 is: {prediction[0]}')
```

#### ***Conclusion***: I have built a regression model that can predict continuous house prices based on two input features: size and location.
